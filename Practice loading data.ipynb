```python
# Import analysis modules
import numpy as np
import xarray as xr
import geopandas
import matplotlib as mpl
import matplotlib.ticker as mticker
import matplotlib.pyplot as plt
from matplotlib import rcParams
import cartopy.crs as ccrs
# Set path to dataset here
filename = "ERA-Int-hgt-00Z.nc"

NC = xr.open_dataset(filename)
NC
```

```python
level = 2
time = np.asarray(NC.time)
t = np.argmax(time)
long = np.asarray(NC.longitude)
lat = np.asarray(NC.latitude)
height = np.asarray(NC.z[t,level,:,:])
plt.pcolormesh(long,lat,height)
```

```python
fig = plt.figure()


ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
A = ax.gridlines(crs=ccrs.PlateCarree(),draw_labels=True,
                linewidth=1, color='gray',alpha=0.1, linestyle='--')
A.top_labels = False
A.right_labels = False

plt.pcolormesh(long,lat,height)

plt.show()
```
